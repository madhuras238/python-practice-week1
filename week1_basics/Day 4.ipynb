```python
#word counter
text = input("Enter a sentence: ")
word = text.split()
print(f"Word count is {len(word)}")
```

```python
#reverse text
text = input("Enter a word: ")
reversed_text = ""
for char in text:
    reversed_text = char + reversed_text
print("Reversed text:", reversed_text)
```

```python
#Replace word
sentence = input("Enter a sentence: ")
print(sentence)
old_word = input("Enter word to be replaced: ")
new_word = input("Enter a new word: ")

updated_sentence = sentence.replace(old_word, new_word)
print(f"Updated sentence: {updated_sentence}")
```

```python
#save user input to file
message = input("Enter a message to save: ")

with open("notes.txt", "w") as file:
    file.write(message)
print("Message saved to notes.txt")
```

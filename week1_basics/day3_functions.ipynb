```python
#Even or Odd number function
def check_even_odd(num):
    if num % 2 == 0:
        return f"{num} is Even"
    else:
        return f"{num} is Odd"

print(check_even_odd(11))
print(check_even_odd(80))
```

```python
#temperature converter function
def cel_to_fah(c):
    fah = (c * 9/5) + 32
    return f"{fah} is temperature in Fahrenheit"

print(cel_to_fah(32))
```

```python
#temperature converter function
def fah_to_cel(fahrenheit):
    celsius = (fahrenheit - 32) * 5/9
    return f"{celsius} is temperature in Celsius"

print(fah_to_cel(90))
```

```python
#count vowels function
def count_vowels(text):
    vowels = "aeiouAEIOU"
    count = 0
    for char in text:
        if char in vowels:
            count = count + 1
    return count

print(count_vowels("Madhura"))
```

```python
#sum of numbers in a list function
def sum_list(numbers):
    total = 0

    for i in numbers:
        total = total + i
    
    return total

print(sum_list([10,25,30]))
```

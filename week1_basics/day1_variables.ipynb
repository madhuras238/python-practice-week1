```python
#Temperature Converter
#Degree celsius to Fahrenheit

temp_c = float(input("Enter temperaure in Degree Celsius: "))
temp_f = (temp_c * 9/5) + 32
print("Temperature in Fahrenheit is: ", temp_f)
```

```python
#emperature Converter
#Fahrenheit to Degree celsius
 
temp_f = float(input("Enter temperature in Fahrenheit: "))
temp_c = (temp_f - 32) * 5/9
print("Temperature in Degree Celsius is: ", temp_c)
```

```python
print("Temperature Converter")
print("1. Degree celsius to Fahrenheit")
print("2. Fahrenheit to Degree celsius")

choice = input("Enter your choice (1 OR 2): ")

if choice == '1':
    temp_c = float(input("Enter temperaure in Degree Celsius: "))
    temp_f = (temp_c * 9/5) + 32
    print("Temperature in Fahrenheit is: ", temp_f)

if choice == '2':
    temp_f = float(input("Enter temperature in Fahrenheit: "))
    temp_c = (temp_f - 32) * 5/9
    print("Temperature in Degree celsius is: ", temp_c)

else:
    print("Invalid choice")
```
